```python
def small_mol_featurizer(df: pd.core.frame.DataFrame = None,
                         mol_obj_col: str = None,
                         featurizer_dict: dict = None) -> pd.core.frame.DataFrame:
    """
    Place Holder
    """
    if df is not None:
        df_with_features = df.copy()
        
        if featurizer_dict is not None:
            for feature in featurizer_dict:
                if mol_obj_col is not None:
                    df_with_features[feature] = df_with_features[mol_obj_col].apply(lambda mol_obj: featurizer_dict[feature](mol_obj))

            return df_with_features
```

```python
featurizer_dict = {'mol_wt':d1.CalcExactMolWt, 
                   'Num_Rings':d1.CalcNumRings, 
                   'Aromatic_rings':d1.CalcNumAromaticRings,
                   'HBA': d1.CalcNumHBA,
                   'HBD': d1.CalcNumHBD,
                   'C_SP3_Hybrid': d1.CalcFractionCSP3,
                   'Num_heavy_atoms': d1.CalcNumHeavyAtoms,
                   'Valence_electrons': d2.NumValenceElectrons}

oled_with_features = small_mol_featurizer(oled_df, 'mol_objs', featurizer_dict)
oled_with_features.head(3)
```

# The RDKit Workflow: Feature Engineering [Structure FingerPrinting]

```python
# Create a function with the Workflow:

def morgan_fp(df: pd.core.frame.DataFrame = None,
              mol_obj_col: str = None,
              num_cir: int = 2,
              fp_size: int = 512) -> pd.core.frame.DataFrame:
    
    """
    Doc String 
    """
    
    fp_df_list = []
    
    if df is not None:
        df2 = df.copy()
              
        if mol_obj_col is not None:
            for mol_obj in df2[mol_obj_col]:
                morg_fp = AllChem.GetMorganFingerprintAsBitVect(mol_obj, num_cir, fp_size)
                morg_fp_string = morg_fp.ToBitString()
                morg_fp_arr = np.fromiter(morg_fp_string, dtype='int')
                fp_df = pd.DataFrame(morg_fp_arr).T
                fp_df_list.append(fp_df)
    
    fp_df = pd.concat(fp_df_list).reset_index(drop=True)
    
    return pd.concat([df2, fp_df], axis=1)  
```

# Data Normalization 

It's often important to ensure your features are on a similar scale, to prevent the magnitude of the values from introducing bias into the model. This is especially important in the case of algorithms that dependend on distances: 

1. k-Nearest Neighbors (KNN) 

* Based on measuring distances between points, feature normalization can prevent 1 feature from dominating the distance calculation.
       
2. Support Vector Machines (SVMs)

* Based on finding the hyperplane to maximize distance between the different classes. If features are on different scales, the hyperplane can be skewed towards the feature with the larger scale. 
        
3. Neural networks (NNs)

* Feature normalization can help prevent vanishing or exploding gradients and speed up convergence. 

# Data Normalization

Some algorithms can make do without normalization:

1. Tree based models (Decision Trees, Random Forests, etc.)

* Make splits based on comparison of individual features, rather than absolute values. 

2. Naive Bayes

* Assumes that features are independent of each other. 


Generally speaking normalization is better for your training even when the algorithms don't require it.

```python
def cif_checker(data: pd.core.frame.DataFrame,
                cif_column: str,
                reduced_formula_column:str,
                output_column_name:str) -> pd.core.frame.DataFrame:
    """
    What this function does:
    
    
    ---
    Parameters
    
    data: Name of the Pandas DataFrame variable.
    cif_column: String for the Pandas DataFrame column containing CIF as string objects.
    reduced_formula_column: String for the Pandas DataFrame column containing the "pretty_formula"/Reduced Formula of the structure. 
    """
    struc_obj_list = []

    for i in range(0,len(data)):
        try:
            struc_obj = IStructure.from_str(data[cif_column][i], fmt='cif')
            if composition.Composition(struc_obj.formula).reduced_formula == data['pretty_formula'][i]:
                struc_obj_list.append(struc_obj)
            else:
                struc_obj_list.append(None)
        except ValueError:
            struc_obj_list.append(None)
    
    data[output_column_name] = struc_obj_list
    
    return data
```

# Features from Structure: JarvisCFID 

```python
# Create a function to apply the featurizer to the DataFrame:

def jarvis_fp(df: pd.core.frame.DataFrame = None,
              validated_structure_col: str = None) -> pd.core.frame.DataFrame:
    
    """
    Doc String 
    """
    
    jarvis = JarvisCFID()
    fp_df_list = []
    
    if df is not None:
        df2 = df.copy()
              
        if validated_structure_col is not None:
            for validated_structure in df2[validated_structure_col]:
                
                jarvis_fp_arr = jarvis.featurize(validated_structure)
                fp_df = pd.DataFrame(jarvis_fp_arr).T
                fp_df_list.append(fp_df)
    
    fp_df = pd.concat(fp_df_list).reset_index(drop=True)
    
    return pd.concat([df2, fp_df], axis=1) 
```

JarvisCFID(use_cell=True, use_chem=True, use_chg=True, use_rdf=True, use_adf=True, use_ddf=True, use_nn=True)

* use_cell (bool): Use structure cell descriptors (4 features, based on DensityFeatures and log volume per atom).

* use_chem (bool): Use chemical composition descriptors (438 features)

* use_chg (bool): Use core charge descriptors (378 features) 

* use_adf (bool): Use angular distribution function (179 features x 2, one set of features for each cutoff).

* use_rdf (bool): Use radial distribution function (100 features)

* use_ddf (bool): Use dihedral angle distribution function (179 features) 

* use_nn (bool): Use nearest neighbors (100 descriptors)

```python
# load the fitted OHE
loaded_ohe = load(open('ohe.pkl', 'rb'))
print(loaded_ohe.categories_)

# Transform the Dataset:
transformed_data = loaded_ohe.transform(inorg_df_form[cols_to_ohe]).toarray()
```

# Features from Formula: Dealing w/ Categorical Data

```python
# Featurizer returns an array, so create a DF from that Array and then Combine with original DataFrame:
# 1. Generate column labels for DataFrame
col_names = [str(value)+'_ohe' for arr in fitted_ohe.categories_ for value in arr]
col_names[0:5]
```

```python
# 2. Generate OneHotEncoded DataFrame:

ohe_df = pd.DataFrame(transformed_data, columns=col_names)
ohe_df.head(3)
```

```python
# 3. Combine w/ original DataFrame
inorg_df_final = pd.concat([inorg_df_form, ohe_df], axis=1)
inorg_df_final.head(2)
```
