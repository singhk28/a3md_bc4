# K Nearest Neighbour

<center><img src="https://github.com/singhk28/a3md_bc4/blob/main/lecture_two/Knn.png?raw=True"/></center>

* One of the simplest algorithms in the ML domain. 

* Distance based algorithm: labels for unseen/new data are predicted based on how close they are to labelled data in the "training" set. 

* Two key parameters: Number of neighbours (k) and how you measure the distance. 


## K Nearest Neighbour: Number of Neighbours

<center><img src="https://github.com/singhk28/a3md_bc4/blob/main/lecture_two/knn_k_val.png?raw=True"/></center>

* As the k value increases the bias in the model decreases (and vice versa):

    * Computational cost increases as k value increases. 
    * Generally k = 3-5 or sqrt(n) (n = number of samples in your training data) are employed. 

<center><img src="https://github.com/singhk28/a3md_bc4/blob/main/lecture_two/knn_loss_curve.png?raw=True"/></center>

* Label is generated via voting (classification) or by taking the average of the closest neighbours (regression).

## K Nearest Neighbour: Strengths & Weaknesses


1. Very easy to understand and gives reasonable performance without lots of tuning.

## Linear Models: Ridge Regression (L2 Regularization)

* Variation of OLS that allows us to control model complexity through regularization. 

* Tries to pick w and b using MSE measurement like OLS, but also makes sure that the w values are as close to 0 as possible by adding a penalty term (minimize the impact of each feature on the predictive outcome). Ridge regression applies L2 regularization:  
    
<center><img src="https://github.com/singhk28/a3md_bc4/blob/main/lecture_two/ridge_pen.png?raw=True"/></center>

* Model complexity (penalty) is controlled in SK-Learn using a hyperparameter called alpha.

    * Higher alpha value = Move weights towards zero. (i.e. more regularized/restricted model).  

## Linear Models: Ridge Regression Demo

```python
from sklearn.linear_model import Ridge

# Initalize model:
ridge_mod = Ridge()

# Fit model with data:
ridge_mod_fitted = ridge_mod.fit(X_train, y_train)

# Generate predictions w/ fitted model
predictions = ridge_mod_fitted.predict(X_test)

# Measure performance:
print('MAE :', mean_absolute_error(y_test, predictions))
```

## Linear Models: Strengths, Weaknesses and Concluding Remarks

* Alpha is the main parameter for linear regression models. We can use it to control model complexity. 
    
    * Greater alpha = simpler model
    
    * Typically vary on a log scale (factor of 10)

* Other key decision is L1 (lasso) regularization vs. L2 (ridge) regularization.

    * Use L1 if you think only a few features are important for predictive value or if you want some insight into which features are important. Otherwise use L2. 
    
* Linear models are fast to train and use for predictions. 

    * scale to large datasets and work well with even sparse datasets.
    
    * Perform well when number of features > samples.
    
    * Often becomes the only option that can be trained reasonably on large datasets. 
    
* Assumes that there is some linear relationship between your target and features. 

* Prone to overfitting and really sensitive to outliers. 

* If you are working in a low-dimensional space, other algorithms may be a better choice.

# Decision Trees

* A popular algorithm built upon extracting mapping principles/rules using a series of if/else questions/statements. 

* Example: Guess what animal I am (bear, hawk, penguin, dolphin):

<center><img src="https://github.com/singhk28/a3md_bc4/blob/main/lecture_two/animal_tree.png?raw=True" style="height: 300px;"/></center>

* The decision tree is an inverted tree with the root at the top connected to the leafs (terminal nodes) through internal/decision nodes and edges (Note: root, internal/decision nodes and leafs are all nodes):

<center><img src="https://github.com/singhk28/a3md_bc4/blob/main/lecture_two/decision_tree_nodes.png?raw=True" style="height: 300px;"/></center>

* The decision criteria (if/else statements such as can fly?) are referred to as 'tests'. 

* Tests can be used for both categorical and continous features. 

* In the case of continous features, the test is a numerical decision boundary:

<center><img src="https://github.com/singhk28/a3md_bc4/blob/main/lecture_two/continous_feature_dt.png?raw=True" style="height: 300px;"/></center>

* The algorithm optimizes what features to use at which nodes and what criteria/question to use in order to get to the answer as quickly as possible. 

* Depending on how many tests we run, we can effectively fit the entire training dataset and get "pure" leafs:

<center><img src="https://github.com/singhk28/a3md_bc4/blob/main/lecture_two/overfitting.png?raw=True" style="height: 300px;"/></center>

* But remember, the goal with ML models is to not memorize (overfit) on our training dataset, but rather to generalize! 

* We we can control model complexity through regularization. 

* Regularization in Decision trees can be achieved through pre or post pruning:
    
    * Pre-pruning: controlling the depth/nodes before creating the model.
    * Post-pruning: building the tree and then removing/collapsing nodes.

* Scikit-learn implements only pre-pruning methods (maximum depth/number of leaves/samples in leaves).

```python
import matplotlib.pyplot as plt

def plot_feature_importance(model, input_data):
    n_features = input_data.shape[1]
    plt.figure(figsize=(16,100))
    plt.barh(range(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), input_data.columns)
    plt.xlabel("Feature importance")
    plt.ylabel("Feature")
    plt.ylim(-1, n_features)
    plt.show()
    
plot_feature_importance(dt_mod_fitted, X_train)
    
```

# Ensemble Methods

* Broadly speaking, ensemble models are models composed of multiple ML models. These can be a mix of different algorithms or the same type of algorithm. We're going to focus on the later (i.e., models made from multiple decision trees). 

* Decision tree ensembles are extremely popular and powerful, especially in terms of tabular data.

* Two approaches to ensemble methods: bagging and boosting.
    1. Bagging: Models are trained independently and you apply some aggregate at the end to get your prediction. 
    2. Boosting: Models are trained in a sequence with each model learning from the previous.

# Decision Tree Ensembles: Random Forests [Bagging]

* Random forests are an ensemble of multiple indepedent decision trees, designed to address the issue of overfitting. 

* The idea is that each tree will be generated in some random manner and have its own bias (overfitting). Granted we average enough of these trees the biases will also average out and reduce overfitting. Generally speaking, we will try to use as many trees as computationally feasible (100s or 1000s) [atleast to the point of threshold].  

* Randomness in each tree is injected through the data points used to build the tree and the features employed for nodes/tests. 

* The number of trees is specified using the n_estimators parameter of the RandomForest method in Sklearn. 

* Each tree will be fed some bootstrapped sample of the input data. [We create a sample the same size as the original dataset but some samples are missing while others are duplicated]. Then each tree will randomly try to select some subset of features (max_features) to create its nodes and tests.

* If we use max_features = n_features then there is no real randomness due to feature select, however, randomness from boostrapping still remains. Using a high value for max_features will generally give you similar trees and they can fit the data with a lower depth. Lower values will require greater depth for the indepdent trees. 

* Random forests can also be analyzed for feature importance, however in this case, the importance is derived via aggregation/averaging of all the trees in the forest.

## Random Forests: Strengths and Weaknesses

1. One of the most widely used algorithms. 

2. Can generally get a decent model without heavy tuning. 

3. Don't need to scale data. 

4. Easy to understand. 

5. Can be computationally expensive for larger datasets. 

6. Tend to perform poorly on high-dimensional sparse data.

7. Remember to fix the random_state otherwise each training will lead to different results! 

# Decision Tree Ensembles: Gradient Boosted Trees [Boosting]

* Build a sequence of (shallow) trees. Each tree is not very deep and isn't expected to learn a lot (weak learner).The  weak learners learn from the mistakes (residuals) of each previous tree to eventually put together a very powerful model. We can control how much each tree learns from the previous tree through the learning_rate. 

    * In addition to the n_estimators, and the low depth, the learning_rate is an important hyperparameter. 
    * Typically tuning on a log scale. 
    * Greater learning_rate leads to a stronger correction and more complex models (higher bias) (i.e. more prone to overfitting).

* Often very good on tabular data and generally lead the way in kaggle competitons. 

* There isn't any randomization by default, it's the strong pruning that prevents the overfitting.

* Because each tree is shallow (typically ~3-5 layers deep), these models are smaller in memory needs and can make fast predictions. 

* There are multiple packages for GradientBoostedTrees ([XGBoost](https://xgboost.readthedocs.io/en/stable/), [CatBoost](https://catboost.ai/en/docs/), [LightGBM](https://lightgbm.readthedocs.io/en/v3.3.2/)).
    * Generally speaking, CatBoost and LightGBM are much faster than XGBoost. XGBoost and LightGBM have a similar performance.

## Boosted Trees: Strengths and Weaknesses

1. One of the most effective algorithms when it comes to tabular data.

2. Hyperparameter tuning and training can be time consuming.

3. Works quite well without normalization.

4. Generally performs poorly on high-dimensional sparse data.

5. In comparison to RF, high n_estimators can lead to a more complex model (i.e. overfitting). Generally, you will pick the number of trees that are compatible with your time and computational resources and then tune the learning rate. 

6. Max_depth is generally advised to be below 5.
